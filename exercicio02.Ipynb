```python
import sqlite3
```

```python
conn = sqlite3.connect('projetofinal.db')
```

```python
import requests
import pandas as pd

url = "https://brasilapi.com.br/api/banks/v1"
req = requests.get(url)
data = req.json()

# Renomear colunas
df = pd.DataFrame(data)
df.rename(columns={'ispb': 'ISPB', 'name': 'Nome', 'code': 'Código', 'fullName': 'Nome Banco'}, inplace=True)#renomer as colunas nome coluna:novo nome

# Tratar  variáveis
df['Código'] = df['Código'].astype('Int64')

# missing
df.fillna({'Código': 0, 'Nome Banco': 'Sem nome'}, inplace=True)  #substituir o NA dos codigos por zeros, tratar nome de bancos sem valor

# Tratamento de  strings
df['Nome'] = df['Nome'].str.upper()  #faz a linha toda nome ficar maiscula
df['Nome Banco'] = df['Nome Banco'].str.title() #faz a primeira linha do nome banco ser maiscula

df.to_sql('df', conn, if_exists='replace', index=False)

df.head(50)
```

```python
import requests
import schedule
import time
import pandas as pd

url = "https://brasilapi.com.br/api/cptec/v1/cidade"

req = requests.get(url)

##req.status_code
##req.json()

data = req.json()


df2 = pd.DataFrame(data)
df2.rename(columns={'nome': 'Cidade','id':'Código', 'estado': 'UF'}, inplace=True)#renomer as colunas nome coluna:novo nome

df2.to_sql('df2', conn, if_exists='replace', index=False)
df2.head(60)
```

```python
import requests
import pandas as pd

url = "https://brasilapi.com.br/api/ibge/uf/v1"
req = requests.get(url)
data = req.json()

df3 = pd.DataFrame(data)

df3['regiao_id'] = df3['regiao'].apply(lambda id: id['id'])
df3['regiao_sigla'] = df3['regiao'].apply(lambda sigla: sigla['sigla'])
df3['regiao_nome'] = df3['regiao'].apply(lambda nome: nome['nome'])

# Dropar a coluna 'regiao' original
df3.drop(columns=['regiao'], inplace=True)
df3.rename(columns={'id': 'Código','sigla':'UF', 'nome': 'Estado'}, inplace=True)#renomer as colunas nome coluna:novo nome

df3.to_sql('df3', conn, if_exists='replace', index=False)
df3.head(27)
```
