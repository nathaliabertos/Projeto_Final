```python
import requests
import pandas as pd

url = "https://brasilapi.com.br/api/ibge/uf/v1"
req = requests.get(url)
data = req.json()

df = pd.DataFrame(data)

df['regiao_id'] = df['regiao'].apply(lambda id: id['id'])
df['regiao_sigla'] = df['regiao'].apply(lambda sigla: sigla['sigla'])
df['regiao_nome'] = df['regiao'].apply(lambda nome: nome['nome'])

# Dropar a coluna 'regiao' original
df.drop(columns=['regiao'], inplace=True)

df.head(27)
```

```python
from datetime import datetime
from plyer import notification

def alerta(erro):
    '''
        Alerta de falha de carregamento de base de dados
    '''
    now = str(datetime.now())

    notification.notify(
            title= 'Aviso de erro',
            message= f'Falha no carregamento da base de dados',
            app_name='Alerta',
            timeout=10
        )

alerta('erro')
```
